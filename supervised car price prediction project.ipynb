```python
V_type_feature=list(set(train_data['vehicleType']))
V_type_feature
```

```python
mean={}
for j in V_type_feature:
    current_mean=X[X.vehicleType==j].price.mean()
    mean[j]=current_mean

def changeVehicleType(str):
    return mean[str]

train_data['VehicleType']=train_data['vehicleType'].apply(changeVehicleType)
test_data['VehicleType']=test_data['vehicleType'].apply(changeVehicleType)

del train_data['vehicleType']
del test_data['vehicleType']
```

```python
train_data.head()
```

```python
brand_feature=set(train_data['brand'])
brand_feature
```

```python
mean={}    
for j in brand_feature:  
    current_mean=X[X.brand==j].price.mean()
    mean[j]=current_mean*0.01

def changeBrand(str):
    return mean[str]

train_data['brand_' + j]=train_data['brand'].apply(changeBrand)
test_data['brand_' + j]=test_data['brand'].apply(changeBrand)

del train_data['brand']
del test_data['brand']
```

```python
model_feature=set(train_data['model'])
model_feature
```

```python
mean={}
for j in model_feature:
    current_mean=X[X.model==j].price.mean()
    mean[j]=current_mean
    
def changeModel(str):
    return mean[str]


train_data['model']=train_data['model'].apply(changeModel)   
test_data['model']=test_data['model'].apply(changeModel)  

del train_data['model']
del test_data['model']
```
